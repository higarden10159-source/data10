```python
#티어 구분해주는 파생변수 추가 - 티어 컬럼
df_pla['tier'] = 'Platinum'
df_dia['tier'] = 'Diamond'
df_ma['tier'] = 'Master'
df_gma['tier'] = 'Grandmaster'
df_chl['tier'] = 'Challenger'
```

```python
#데이터 수직 결합 concat
df_tier = pd.concat([df_pla, df_dia, df_ma, df_gma, df_chl], axis=0, ignore_index=True, join='inner')
```

```python
df_tier
```

```python
#이상치 1 게임 아이디 체크
gameId_count = df_tier.groupby('gameId')['gameId'].count()
gameId_count
```

```python
abnormal_games = gameId_count[gameId_count != 8]
print("정상 게임 수:", gameId_count[gameId_count == 8].count())
print("비정상 게임 수:", abnormal_games.count())
```
